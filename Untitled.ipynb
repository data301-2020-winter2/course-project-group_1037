```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid',font_scale=2)

df = sns.load_dataset("penguins")
```

```python
plt.figure(figsize=(20,5))
vp = sns.violinplot(data=df,x='species',y='bill_length_mm')
plt.xlabel('Species')
plt.ylabel('Bill Length')
plt.title('Penguin Bill Length by species')
sns.despine()
vp.get_figure().savefig('parsons1.png')

```

```python
sns.violinplot(y = 'bill_length_mm', x = 'species', data = df, hue='sex', split=True).set(xlabel="Species",ylabel="Bill Length")
```
